```python
from transformers import AutoTokenizer
import pandas as pd
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")   # getting the tokenizer for training it
```

```python
# training on new corpus, here i am right now using just two sentences as corpus
tokenizer.train_new_from_iterator(["hello world", "goodbye world"], vocab_size=50000)
```

```python
# After training the tokenizer on custom dataset, tokenizing the new sentence using that new tokenizer and feeding that into the bert transformer

import tensorflow as tf
from transformers import TFAutoModelForSequenceClassification


# Tokenize the input sentence
sentence = "Dilip pokhrel is amazing"
tokens = tokenizer(sentence, padding=True, truncation=True, return_tensors="tf")
print(tokens.tokens())
# Load the pre-trained BERT model for sequence classification
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-cased")

# Forward pass through the model
logits = model(tokens.input_ids, attention_mask=tokens.attention_mask)[0]

# Get the predicted class (assuming binary classification)
predicted_class = tf.argmax(logits, axis=1).numpy()[0]

# Print the logit score and predicted class
print("Logit score:", logits[0][predicted_class].numpy())
print("Predicted class:", predicted_class)
```
